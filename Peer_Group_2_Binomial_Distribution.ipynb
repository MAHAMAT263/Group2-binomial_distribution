# **Example: Defective Products**

**Scenario:** A factory produces batteries, where 3% of them are defective. If 100 batteries are produced, what is the probability that exactly 4 batteries are defective?

Plot in X-Y Here is the Python code to generate the plot with labeled X and Y axes:

**Python code example 1 implemented with numpy only**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 100  # Number of trials
p = 0.03 # Probability of success

# X values (number of successes)
x = np.arange(0, 16)

# Binomial probability mass function (PMF) using numpy
pmf = [(np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k))) * (p**k) * ((1 - p)**(n - k)) for k in x]
y = pmf
# Create the plot
plt.figure(figsize=(8, 6))
plt.bar(x, y, color='skyblue', alpha=0.7)
plt.plot(x, y, 'bo-', label='Binomial PMF')  # X-Y plot overlaid
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.legend()
plt.show()
```
